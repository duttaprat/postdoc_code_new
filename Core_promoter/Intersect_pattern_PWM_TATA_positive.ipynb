```python
# Convert dataframes to BED format
# Convert dataframes to BED format including TSS column
def df_to_bed(df, chrom_col, start_col, end_col, extra_cols=[]):
    cols = [chrom_col, start_col, end_col] + extra_cols
    return pybedtools.BedTool.from_dataframe(df[cols])
```

```python
# Create BEDTools objects
bed1 = df_to_bed(df_TATA_PWM, 'Chromosome', '45BP_Start', '45BP_End', ['TSS', 'Sequence'])
bed2 = df_to_bed(df_TATA_pattern, 'Chromosome', '45BP_Start', '45BP_End', ['TSS'])
```

```python
# Find intersections
intersections = bed1.intersect(bed2, wa=True, wb=True)
```

```python
# Convert the intersections to a DataFrame
intersected_df = intersections.to_dataframe(names=[
    'Chromosome', 'Start1', 'End1', 'TSS1', 'Sequence', 'Chromosome2', 'Start2', 'End2', 'TSS2'
])
intersected_df
```

```python
# Calculate the overlap percentage
def calculate_overlap_percentage(row):
    overlap_start = max(row['Start1'], row['Start2'])
    overlap_end = min(row['End1'], row['End2'])
    overlap = max(0, overlap_end - overlap_start + 1)
    length1 = row['End1'] - row['Start1'] + 1
    length2 = row['End2'] - row['Start2'] + 1
    return overlap / min(length1, length2)

intersected_df['Overlap_Percentage'] = intersected_df.apply(calculate_overlap_percentage, axis=1)
intersected_df
```

```python
full_overlap_sequences = intersected_df[intersected_df['Overlap_Percentage'] == 1.0][['Chromosome', 'Start1', 'End1', 'TSS1', 'Sequence']]
full_overlap_sequences
```

```python
full_overlap_sequences.to_csv("/data/projects/CorePromoterPWM/Intersect_PWM_pattern/TATA_positive.tsv", sep="\t", index=False)
```
