```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
df_positive = pd.read_csv("/data/private/pdutta/DNABERT_data/Core_Prom_new/TATA/raw_positive.tsv", sep="\t")
df_negative = pd.read_csv("/data/private/pdutta/DNABERT_data/Core_Prom_new/TATA/raw_negative.tsv", sep="\t")
```

```python
df_positive
```

```python
df_TATA = df_positive[['Chromosome','50BP_Start','50BP_End','Sequence']]
df_TATA = df_TATA.rename(columns={
    '50BP_Start':'Start',
    '50BP_End':'End'
})
df_TATA['Label'] = 1
df_TATA
```

```python
df_TATA_negative = df_negative[['Chromosome','Start','End','Sequence']]
df_TATA_negative['Label'] = 0
df_TATA_negative
```

```python
def seq2kmer(seq):
    """
    Convert original sequence to kmers
    
    Arguments:
    seq -- str, original sequence.
    k -- int, kmer of length k specified.
    
    Returns:
    kmers -- str, kmers separated by space
    """
    k=6
    kmer = [seq[x:x+k] for x in range(len(seq)+1-k)]
    kmers = " ".join(kmer)
    return kmers
```

```python
df= pd.concat([df_TATA, df_TATA_negative]).reset_index(drop=True)
df = df.rename(columns={
    'Sequence':'Reference_sequence'
})
```

```python
df['Sequence'] = df['Reference_sequence'].apply(seq2kmer)
```

```python
df
```

```python
train, test = train_test_split(df, test_size=0.20)
```

```python
train
```

```python
output_path = "/data/private/pdutta/DNABERT_data/Core_Prom_new/TATA"
```

```python
train.to_csv(output_path+"/raw_train.tsv", sep="\t", index=False)
test.to_csv(output_path+"/raw_test.tsv", sep="\t", index=False)
```

```python
train[['Sequence','Label']].to_csv(output_path+ "/train.tsv", sep="\t", index=False)
test[['Sequence','Label']].to_csv(output_path+ "/dev.tsv", sep="\t", index=False)
```
