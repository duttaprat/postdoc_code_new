```python
df_TATA_negative_pattern= pd.read_csv("/data/private/pdutta_new/non_coding_regions/TATA_NonTATA/Pattern/TATA_raw_negative_all.tsv", sep="\t")
df_TATA_negative_pattern
```

```python
def df_to_bed(df, chrom_col, start_col, end_col, extra_cols=[]):
    cols = [chrom_col, start_col, end_col] + extra_cols
    return pybedtools.BedTool.from_dataframe(df[cols])
```

```python
# Create BEDTools objects
bed1 = df_to_bed(negative_tata_core_df, 'Chromosome', '45BP_Start', '45BP_End', ['Relative_Score','Position', 'Sequence'])
bed2 = df_to_bed(df_TATA_negative_pattern, 'Chromosome', 'Start', 'End',['TATA_Box_Position'])
```

```python
# Find intersections
intersections = bed1.intersect(bed2, wa=True, wb=True)
```

```python
# Convert the intersections to a DataFrame
intersected_df = intersections.to_dataframe(names=[
    'Chromosome', '45BP_Start', '45BP_End', 'Relative_score', 'Position', 'Sequence', 'Chromosome2', 'Start', 'End', 'TATA_Box_Position'
])
intersected_df
```

```python
full_overlap = intersected_df[intersected_df['45BP_Start']==intersected_df['Start']]
full_overlap
```

```python
full_overlap.to_csv("/data/projects/CorePromoterPWM/Intersect_PWM_pattern/TATA_negative.tsv", sep="\t", index=False)
```

```python
full_overlap_sequences = intersected_df[intersected_df['Overlap_Percentage'] == 1.0][['Chromosome', 'Start1', 'End1', 'TSS1', 'Sequence']]
full_overlap_sequences
```

```python
full_overlap_sequences.to_csv("/data/projects/CorePromoterPWM/Intersect_PWM_pattern/TATA_positive.tsv", sep="\t", index=False)
```
